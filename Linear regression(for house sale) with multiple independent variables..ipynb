# #VIF

```python
# Removing variables with multicollinearity by VIF.
#Calculating VIF
#importing variance_inflation_factor function from Statsmodels.

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_df = X

#Calculating VIF for every I.V column.

VIF = pd.Series([variance_inflation_factor(vif_df.values,i) for i in range (vif_df.shape[1])], index=vif_df.columns)
VIF
```

Remove variables having VIF>5 above. We have total 7 such variables.

```python
#Checking the index where max value of VIF is stored.

VIF[VIF== VIF.max()].index[0]
```

```python
#Remove this index having max VIF.

def MC_remover(df):
    vif=pd.Series([variance_inflation_factor(df.values, i)for i in range (df.shape[1])], index=df.columns)
    if vif.max () >5:
        print(vif[vif==vif.max()].index[0], 'has been removed')
        df=df.drop(columns=[vif[vif==vif.max()].index[0]])
        return df
    else:
        print('no multicollinearity present anymore')
        return df
```

```python
#Estimated above 7 I.V causing multicollinearity should be removed.

for i in range(7):
    vif_df=MC_remover(vif_df)
```

# Adjusted R2

```python
#Forward feature selection
def adj_r2 (df_X, df_Y, predictions):
    from sklearn.metrics import r2_score 
    R= r2_score (df_Y, predictions)
    n=len(df_Y)
    m=len(df_X.columns)
    adj_R=1- ((1-R)*(n-1))/(n-m-1)
    return adj_R
    
```

```python
adj_r2(X_test, Y_test, predictions)
```

Compare it with R2..Adj_r2(0.8455) < R2(0.8461) it means there are some variables which are not adding value to the final model.

```python
def return_r2 (X, Y, model):
    model.fit(X, Y)
    pred= model.predict(X)
    m=len(X.columns)
    r2=adj_r2(m, Y, pred)
    return r2

```
